```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
sep_coins = cv2.imread('C:/Users/james/Desktop/udemy/UPDATED_NLP_COURSE/Computer-Vision-with-Python/DATA/pennies.jpg')
```

```python
display(sep_coins)
```

```python
#we apply 'median blur' to the image, medium blur helps me to get read of features
#I really don't need, such as the faces, so we can focus on circles themselves
#Then I convert it in "Gray scale" 
#apply a "binary threshold" in order to make it black and white 
#then find the contours
```

```python
sep_blur = cv2.medianBlur(sep_coins,25)
```

```python
#display the result of that blur
display(sep_blur)
```

```python
#convert in gray scale
gray_sep_coins = cv2.cvtColor(sep_blur,cv2.COLOR_BGR2GRAY)
```

```python
display(gray_sep_coins)
```

```python
#binary threshold ( background and foreground)
ret, sep_thres = cv2.threshold(gray_sep_coins,160,255,cv2.THRESH_BINARY_INV)
```

```python
display(sep_thres)
```

```python
#now find the contours
image,contours,hierarchy = cv2.findContours(sep_thres.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
#
for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(sep_coins,contours,i,(255,0,0),10)
    
```
